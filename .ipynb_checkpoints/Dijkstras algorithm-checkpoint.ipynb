```python
import numpy as np
import random


m = int(input("Enter the number of nodes: "))

matrix = np.zeros((m,m))
# matrix = np.array([[-1,6,0,1,0],[6,-1,5,2,2],[0,5,-1,0,5],[1,2,0,-1,1],[0,2,5,1,-1]])
matrix = np.array([[-1,4,2,0,0],[4,-1,3,2,3],[2,1,-1,4,5],[0,2,4,-1,1],[0,3,5,1,-1]])
  
# # For user input 
# for i in range(m):          # A for loop for row entries 
#     for j in range(m):      # A for loop for column entries
#         if i==j:
#             matrix[i,j] = -1
#         else:
#             matrix[i,j] = matrix[j,i] = random.randint(0,5)
        
print(matrix)        
source = int(input("Enter the source: "))
target = int(input("Enter the target: "))


dist = np.zeros(m)
prev_node = []
visited = []
prev_dist = 0
count = 0

for i in range(len(dist)):
    if i == source:
        dist[i] = 0
    else:
        dist[i] = float('inf')
print("Initial",dist)



while len(visited) != m:
    print("----------------------------------------------------")
    print("while_count",count)
    
    print("visited_update",visited)
    
    
    # 1. For picking the appropriate row
    if count == 0:
        row_no = source         #Index of the row to be picked
        row = []                #Row of the source node
        for i in range(m):      #Loop for picking the row corresponding to the node
            row.append(matrix[row_no,i])    
        print("main_if",row_no)
    else:
        
        dist_temp = np.zeros(m)         #  Temporary distance array for comparision
        for i in range(len(dist)):
            
            if dist[i] == 0:             #  Makes the distance element infinity if it is zero
                dist_temp[i] = float('inf')
            else:                        #  Makes the distance element infinity if it is previous node 
                dist_temp[i] = dist[i]
            
            for j in range(len(visited)):           #  To make previous node position infinity
                
                if i == visited[j]:
                    dist_temp[i] = float('inf')
                    
            if i == row_no:              #  To make current node distance infinity
                dist_temp[i] = float('inf')

        
            
            
            
        smallest = dist_temp[0]
        for i in range(m):  # To find the smallest distance to pick the next node
            if smallest >= dist_temp[i]:
                smallest = dist_temp[i]
                smallest_position = i

        row_no = smallest_position
        row = []
        for i in range(m):      #Loop for picking the row corresponding to the node
            row.append(matrix[row_no,i]) 
        print("Picked _row_no",row_no)
        print("Picked _row",row)
        

    #_________________________________________________________________________________________#
    
    # 2. For adding the previous distance to get distance form the start node
    if count > 0:
        for i in range(m):
            if i == visited[count-1]:
                prev_dist = prev_dist + row[i]
            else:
                prev_dist = prev_dist
            
    #_________________________________________________________________________________________#
            
    # 3. Adding the previous distance to the row array for comparision
    
    
    comp = np.zeros(m)   # Array for comparision (a array in my algorithm)
    for i in range(m):
        if row[i] > 0:
            comp[i] = row[i] + prev_dist
            
        else:
            comp[i] = row[i]
            
            
            
    
    # 4. For comparing the comp array and dist array to get new distance
    
    for i in range(m):
        if comp[i]>0  and comp[i]<dist[i]:
            dist[i] = comp[i]
        
        else:
            dist[i] = dist[i]
            
    visited.append(row_no)
        
    count+=1
    print(dist)

            
            
```
